# Day 12 / 30 — Streamlit live demo

**Project:** 6-class news article categorisation  
**Stack:** Streamlit · same `load_artifact` / `predict_with_details` as Day 11  
**Prerequisite:** `models/news_hybrid.joblib` from `uv run python train.py`

---

## What we ship

| Piece | Role |
|-------|------|
| `app.py` | Browser UI: paste text, **Classify**, view predicted label + LinearSVC decision scores |
| `src/inference.py` | `predict_with_details()` — label + per-class `decision_function` scores for charts |

The app caches the joblib bundle with `@st.cache_resource` so restarts stay fast after the first load.

## Run locally

From the **repository root** (not `notebooks/`):

```bash
uv run streamlit run app.py
```

Then open the URL Streamlit prints (usually `http://localhost:8501`).

**Tip:** Use the sidebar to load example blurbs per category, then click **Classify**.
